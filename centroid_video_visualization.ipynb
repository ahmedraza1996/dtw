```python
import cv2
import os
```

```python
source_video_path = "/home/ahmed/Ahmed_data/agglomerative/data/2020-12-1817-12-49.mp4"
src_cluster_path ='/home/ahmed/Ahmed_data/agglomerative/clusters/clusters'
tgt_cluster_path ='/home/ahmed/Ahmed_data/agglomerative/clusters/clusters_target' 
output_folder = "/home/ahmed/Ahmed_data/agglomerative/output/2020-12-1817-12-49.avi"
image_folder = '/home/ahmed/Ahmed_data/agglomerative/src_frames'
```

```python
# Define the size of the video window and image
video_window_size = (640, 480)
image_size = (640, 480)
```

```python
def read_clip_indices_from_txt(file_path):
    clip_indices = {}
    with open(file_path, 'r') as file:
        for line in file:
            clip_name, indices = line.strip().split(':')
            start_idx, end_idx = map(int, indices.split(','))
            clip_indices[int(clip_name)] = (start_idx, end_idx)
    return clip_indices
```

```python
def find_clip_key(clip_indices, frame_index):
    for key, (start_idx, end_idx) in clip_indices.items():
        if start_idx <= frame_index <= end_idx:
            return key
    return None 
```

```python
src_indices =read_clip_indices_from_txt(src_cluster_path)
tgt_indices =read_clip_indices_from_txt(tgt_cluster_path)
```

```python
video_capture = cv2.VideoCapture(source_video_path)


# Get video properties

frame_width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_rate = int(video_capture.get(cv2.CAP_PROP_FPS))
#fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
#out = cv2.VideoWriter(output_folder,
#                                     fourcc,
#                                     frame_rate, (frame_width *2, frame_height*2))

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_folder, fourcc, frame_rate, (1280,480))

frame_count = 0 
while True:
    ret, frame = video_capture.read()

    if not ret:
        break
    resized_frame = cv2.resize(frame, video_window_size)
    tgt_index = find_clip_key(tgt_indices, frame_count)
    print(tgt_index)
    start_idx_tgt, end_idx_tgt= tgt_indices.get(tgt_index)
    start_idx, end_idx= src_indices.get(tgt_index)
    mid = int((start_idx+ end_idx)/2)
    image_path = os.path.join(image_folder, f"{mid}.jpg")
    image = cv2.imread(image_path)
    resized_image = cv2.resize(image, image_size)
    
    timeline_tgt = f"Frame: {frame_count} Clip {tgt_index}: {start_idx_tgt} - {end_idx_tgt}"
    timeline_frame = cv2.putText(resized_frame, timeline_tgt, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    timeline = f"Clip {tgt_index}: {start_idx} - {end_idx}"
    resized_image = cv2.putText(resized_image, timeline, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    combined_frame = cv2.hconcat([resized_image, timeline_frame])        
    # Show the video and image side by side
    cv2.imshow('Video with Image', combined_frame)
    cv2.waitKey(1)  # Show the frame for a short time (1ms)

    # Write the frame to the output video
    out.write(combined_frame)
    # Pause between clips (1s)
    
    frame_count+=1
    # Release the video capture and writer objects
video_capture.release()
out.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```
